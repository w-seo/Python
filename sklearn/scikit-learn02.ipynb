```python
# 線形モデルを準備
from sklearn import linear_model

# 識別機を作成
clf = linear_model.LogisticRegression()
```

```python
clf
```

```python
# データの個数
n_samples = X.shape[0]
# 半分のデータを学習
n_train = n_samples // 2
# テストデータ数
n_test = n_samples - n_train
```

```python
# 0,1,...,n_train-1 : 最初の半分
train_index = range(0, n_train)

# n_train, n_train+1,...n_samples-1 : 残りの半分
test_index = range(n_train, n_samples)
```

```python
# 確認してみる
np.array(train_index), np.array(test_index)
```

```python
# テストデータ
X_test = X[test_index]
# 学習データ
X_train = X[train_index]

# テストデータのラベル
y_test = y[test_index]
# 学習データのラベル
y_train = y[train_index]
```

```python
# 識別器の学習
clf.fit(X_train, y_train)
```

```python
# 学習データの精度
print(clf.score(X_train, y_train)) 
```

```python
# テストデータの精度
print(clf.score(X_test, y_test))
```

```python
# テストデータの識別
clf.predict(X_test)
```

```python
wrong = 0

for i, j in zip(clf.predict(X_test), y_test):
    if i == j:
        print(i, j)
    else:
        print(i, j, " Wrong!")
        wrong += 1
```

```python
print("{0} / {1} = {2}".format(wrong,
                               n_test, 
                               1 - wrong / n_test ))
```
